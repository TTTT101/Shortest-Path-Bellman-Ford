```python
import numpy as np
import xlsxwriter

np.random.seed(4) # use your last 2 digits of GTID
########## THIS IS INPUT CREATION, DO NOT MODIFY ANY ENTRY
number_of_nodes = 25 # nodes will be numbered from 0 to 24, 
number_of_dimensions = 4 
size_box=50 
list_of_coordinates=[]
for i in range(0,number_of_nodes):
    dimensions=np.random.randint(0,size_box+1,size=number_of_dimensions)
    list_of_coordinates.append(dimensions)
    
shipping_costs = np.random.randint(0,size_box*(number_of_dimensions**2),
                              size = (number_of_nodes,number_of_nodes))
for i in range(0,number_of_nodes):
    shipping_costs[i,i]=0
l1_distances=np.zeros(shape = (number_of_nodes,number_of_nodes)) 
lmax_distances =np.zeros(shape = (number_of_nodes,number_of_nodes))
```

```python

'''
Here you compute the l1 and lmax distances
'''
l1_distances

for i in range(len(list_of_coordinates)):
    for j in range(len(list_of_coordinates)):
        #print(list_of_coordinates[i], (list_of_coordinates[j]), abs(list_of_coordinates[i]-list_of_coordinates[j]), abs(list_of_coordinates[i]-list_of_coordinates[j]).sum())
        #print(abs(list_of_coordinates[i]-list_of_coordinates[j]).sum())
        distl1 = abs(list_of_coordinates[i]-list_of_coordinates[j]).sum()
        l1_distances[i][j] = distl1
        distl_infinity = max(abs(list_of_coordinates[i]-list_of_coordinates[j]))
        lmax_distances[i][j] = distl_infinity

#l1_distances
#lmax_distances
```

```python
cost_matrix = shipping_costs+l1_distances ### for part a 
cost_matrix = shipping_costs+lmax_distances ### for part b
# cost_matrix[i,j] is the weight of i,j arc
#you can also manually create the cost_matrix 
```

```python
### Bellman-Ford algorithm ### version 2
cost_matrix = shipping_costs+l1_distances ### for part a 
N=25 # number of nodes
arc_pairs_list =[]
for i in range(N):
    for j in range(N):
            arc_pairs_list.append([i,j])
```

```python
previous_best_i_to_t  = np.zeros(shape=(1,N))
previous_best_i_to_t=previous_best_i_to_t+1e10  #really big cost for initial

previous_best_i_to_t[0,N-1]=0
previous_best_i_to_t
```

```python
arc_jumps_to_minimize_node_cost = []  #node arc pair

for k in range(N):
    print('k = ', k)
    print('######################################## k  ############')
    for node in -np.arange(start=-24, stop=1):

        possible_arcs = [x if x[0]==node else np.nan for x in arc_pairs_list]

        possible_arcs_filtered=[]
        #get possible arcs 
        for i in possible_arcs:
            try :
                len(i)
                possible_arcs_filtered.append(i)
            except :
                pass

        print('possible arc filtered : ',possible_arcs_filtered)
        print(previous_best_i_to_t[0,node])

        #get cost of arcs 
        all_costs = np.array([cost_matrix[x[0],x[1]] for x in possible_arcs_filtered])
        all_previous_costs = np.array([previous_best_i_to_t[0,x[1]] for x in possible_arcs_filtered])

        print('All cij : ',all_costs)
        print('All previous yjk-1 costs :',all_previous_costs)

        print(min(all_costs+all_previous_costs))
        index_min_sum = np.argmin(all_costs+all_previous_costs)
        print('min cost item arc sum term',index_min_sum)

        print('min arc yileding for this node ', possible_arcs_filtered[index_min_sum])


        print(min(min(all_costs+all_previous_costs) , previous_best_i_to_t[0,node]))   # complex yki function
        new_best_cost_node = min(min(all_costs+all_previous_costs) , previous_best_i_to_t[0,node])  #full complex yki function
        
        if new_best_cost_node<previous_best_i_to_t[0,node]:
        
            previous_best_i_to_t[0,node] = new_best_cost_node
            arc_jumps_to_minimize_node_cost.append([k,node,possible_arcs_filtered[index_min_sum]])
        else:
            print('no improvement in this run of k({}) for this node {}:'.format(k, node))
        #record for tracking the best path
        
        #record index sum if and only if the min of sum is smaller than that of yik-1
        #previous step 
        
        
        
        
        #cost components of yik-1 for each node
        print( 'yik-1  for each node : ',previous_best_i_to_t)

    #     break
```

```python
#find the path
array_arc_jumps_to_minimize_node_cost=np.array(arc_jumps_to_minimize_node_cost, dtype=object)

final_map = []
for node in range(N-1):
    
    node_item_entry=[]
    for i in range(len(array_arc_jumps_to_minimize_node_cost)):
        if array_arc_jumps_to_minimize_node_cost[i][1]==node:
            node_item_entry.append(array_arc_jumps_to_minimize_node_cost[i])
    
    #for this node get most recent update
    
    final_map.append([node_item_entry[-1]])
```

```python
final_map
```

```python
for node in range(N-1):
    path = []
    node_start=node
    #print(node)
    
    for i in range(N):
        path.append(node)
        
        next_node=final_map[node][0][2][1]
        node=next_node
        if  node==N-1:
            path.append(N-1)
            break
    
    print('SP distance between node {} and {} is : {} and one SP is : {}'.format(node_start, N-1,
                                                                              previous_best_i_to_t[0][node_start], 
                                                                              path))
    
```

```python
########## These lines of code allow you to export the matrix to an excel file if needed
workbook = xlsxwriter.Workbook('cost_matrix.xlsx')
worksheet = workbook.add_worksheet()
for i in range(0,number_of_nodes):
    worksheet.write(i,i,cost_matrix[i,i])
    for j in range(i+1,number_of_nodes):
        worksheet.write(i, j, cost_matrix[i,j] )
        worksheet.write(j, i, cost_matrix[j,i] )
workbook.close()
```
